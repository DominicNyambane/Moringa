```python
import pandas as pd
import numpy as np
```

```python
# Empty dataframe of 20 rows
data = pd. DataFrame(index=range(20))
```

```python
# populate column A with names randomly
data["A"] = np.random.choice(["Eric", "Lumumba", "Chemu"], size=len(data))
# populate column B with random integers
data["D"] = np.random.randint(10, 20, size=len(data))
# populate column C with random floats
data["C"] = np.random.random(size=len(data))
```

```python
data.head()
```

```python
# Empty dataframe of 20 rows
add_data = pd. DataFrame(index=range(10))
```

```python
np.random.seed(43)
# populate column A with names randomly
add_data["A"] = np.random.choice(["Kirui", "Ida", "Maggy"], size=len(add_data))
# populate column B with random integers
add_data["D"] = np.random.randint(10, 20, size=len(add_data))
# populate column C with random floats
add_data["C"] = np.random.random(size=len(add_data))
```

```python
add_data.head()
```

```python
new_data = pd.concat([data, add_data], axis=0, ignore_index=True)
```
